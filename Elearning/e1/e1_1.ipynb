2. Các kiểu dữ liệu cơ bản và phép toán trên chúng (string, interger, float, bool)

```python
while True:
    print("Chọn phép tính \n"
          "1: Cộng\n"
          "2: Trừ\n"
          "3: Nhân\n"
          "4: Chia")

    chose = int(input("Chọn từ 1-4: "))

    if chose < 1 or chose > 4:
        print("Lựa chọn không hợp lệ! Chọn lại.\n")
        continue

    print("Nhập số:")
    a = eval(input("Nhập a: "))
    b = eval(input("Nhập b: "))

    if chose == 1:
        c = a + b
        print(f"{a} + {b} = {c} Kiểu dữ liệu: {type(c)}")
    elif chose == 2:
        c = a - b
        print(f"{a} - {b} = {c}  Kiểu dữ liệu: {type(c)}")
    elif chose == 3:
        c = a * b
        print(f"{a} * {b} = {c} Kiểu dữ liệu: {type(c)}")
    elif chose == 4:
        if b == 0:
            print("Không thể chia cho 0!")
        else:
            c = a / b
            print(f"{a} / {b} = {c}  Kiểu dữ liệu: {type(c)}")

    # Hỏi người dùng có tiếp tục không
    tiep = input("Bạn muốn thực hiện phép tính khong? (y/n): ")
    if tiep.lower() != "y":
        print("Kết thúc chương trình")
        break
```

3. Các kiểu dữ liệu nâng cao: List, Tupble, Set, Dictionary

list : thay đổi được, có thứ tự , cho phép trùng lập , chứa nhiều kiểu

(có thể sửa thêm xóa phàn tử)

```python
numbers = [1,2,3,4,5,6,7]
print(numbers[0])
numbers.append(9)
numbers[1]=10
print(numbers)
```

Tuple: không thay đổi được, có thứ tự, trùng lập 
(Khi dữ liệu không được phép sửa)

```python
point = (10, 20)
print(point[0])   # 10

# point[0] = 99 ->  lỗi k sửa được
```

Set  : khong có thứ tự , khong trùng, tự động loại trùng thay đỏi được
Dùng để: 
dùng loại bỏ trùng lặp 
Khi cần các phép toán tập hợp: giao, hợp, hiệu…

```python
my_set = {1, 2, 2, 3, 4}
print(my_set)     # {1, 2, 3, 4}

# Thêm phần tử
my_set.add(10)

# Phép toán tập hợp
a = {1, 2, 3}
b = {3, 4, 5}
print(a & b)  # giao -> {3}
print(a | b)  # hợp
```

Dictionary (Từ điển / Map)
Lưu trữ theo kieur key-value
thay đổi được 
Khi dùng?
Khi bạn muốn dữ liệu có nhãn hoặc tên, không phải chỉ số
Ví dụ: thông tin sinh viên, cấu hình, đối tượng…

```python
student = {
    "name": "Hân",
    "age": 19,
    "major": "IT"
}

print(student["name"])  

student["age"] = 20       # sửa
student["score"] = 9.0    # thêm
print(student)
```

Mod
"r"	Chỉ đọc (read), file phải tồn tại
