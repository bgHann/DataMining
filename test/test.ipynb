```python
#mở file 
f = open("doc.txt ", "r")
print(f.read()) #real(nub)=> trả về số char đàu tiên
f.close() #luon đóng tệp khi hoàn tất
```

```python
with open("doc.txt") as f :
    print(f.read()) #with sẽ tự đóng file
```

readline() đọc dòng

```python
with open("doc.txt") as f: 
   for x in f : #lặp tệp từng dòng
    print(x)
```

Ghi vào tệp hiện có
Để ghi vào một tệp tin hiện có, bạn phải thêm một tham số vào open()hàm:

"a"- Thêm - sẽ thêm vào cuối tệp

"w"- Viết - sẽ ghi đè lên bất kỳ nội dung hiện có nào

```python
with open("doc.txt", "a") as f :
    f.write("\n nội dung them vào")
with open("doc.txt") as f :
    print(f.read())
```

```python
with open("doc.txt", "w") as f:
    f.write("nội dung chèn")
with open("doc.txt") as f:
    print(f.read())
```

Tạo tệp mới
Để tạo một tệp mới trong Python, hãy sử dụng open()phương thức này với một trong các tham số sau:

"x"- Tạo - sẽ tạo một tệp, trả về lỗi nếu tệp tồn tại

"a"- Thêm - sẽ tạo một tệp nếu tệp được chỉ định không tồn tại

"w"- Write - sẽ tạo một tập tin nếu tập tin được chỉ định không tồn tại
