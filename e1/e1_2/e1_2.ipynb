```python
# -----------------------------
# ELEARNING2.2
# -----------------------------
# BÙI GIA HÂN_064305011558

import os
import re
from collections import Counter

#  Đường dẫn file
file_path = r"D:\GIAHAN\CN2304CLCA\IT_Major\khaikhacdulieu\DataMining\e1\e1_2\SMSSpamCollection"

# Kiểm tra file có tồn tại không
if not os.path.exists(file_path):
    raise FileNotFoundError(f"File không tồn tại: {file_path}")

# Đọc file
with open(file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

print("Tổng số dòng trong file:", len(lines))
print("Dòng đầu tiên:", lines[0])

#  Đếm số dòng spam và ham
spam_count = 0
ham_count = 0

for line in lines:
    parts = line.split("\t")
    if len(parts) < 2:
        continue  # bỏ dòng trống hoặc không hợp lệ
    label = parts[0].lower()
    if label == "spam":
        spam_count += 1
    elif label == "ham":
        ham_count += 1

print("\nSố dòng spam:", spam_count)
print("Số dòng ham:", ham_count)

#  Đếm tổng số từ (không phải số)
word_count = 0
for line in lines:
    parts = line.split("\t")
    if len(parts) < 2:
        continue
    text = parts[1]
    words = text.split()
    for w in words:
        if not w.isdigit():
            word_count += 1

print("\nTổng số từ (không phải số):", word_count)

#  Đếm số từ 'you' (không phân biệt hoa thường)
you_count = 0
for line in lines:
    parts = line.split("\t")
    if len(parts) < 2:
        continue
    text = parts[1].lower()
    words = text.split()
    you_count += words.count("you")

print("\nSố từ 'you':", you_count)

#  Đếm số từ khác nhau
unique_words = set()
for line in lines:
    parts = line.split("\t")
    if len(parts) < 2:
        continue
    text = parts[1].lower()
    words = re.findall(r'\b[a-zA-Z]+\b', text)
    unique_words.update(words)

print("\nSố từ khác nhau:", len(unique_words))

#  Liệt kê 10 từ xuất hiện nhiều nhất
all_words = []
for line in lines:
    parts = line.split("\t")
    if len(parts) < 2:
        continue
    text = parts[1].lower()
    words = re.findall(r'\b[a-zA-Z]+\b', text)
    all_words.extend(words)

counter = Counter(all_words)
most_common = counter.most_common(10)

print("\n10 từ xuất hiện nhiều nhất:")
for word, freq in most_common:
    print(word, ":", freq)
```
